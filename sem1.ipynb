# Семинар 1. Введение в Pandas

На этом семинаре мы познакомимся с библиотекой `pandas`. Она является фактическим стандартом для работы с табличными данными в Python.

Для начала импортируем необходимые библиотеки:

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

## 1. Загрузка данных

Мы будем работать с датасетом Spotify. Функция `pd.read_csv()` — это основной метод для чтения текстовых файлов с разделителями (обычно запятыми).

Она может принимать огромное количество параметров (например, `sep` для указания разделителя, `index_col` для выбора колонки-индекса), но базово ей достаточно передать путь к файлу или прямую ссылку на него. Pandas сам скачает данные по URL и преобразует их в объект `DataFrame` — двумерную таблицу, похожую на лист Excel.

```python
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv"
df = pd.read_csv(url)
```

### Метод `.head()` и `.tail()`

Чтобы не выводить всю таблицу (которая может содержать миллионы строк) и не перегружать память, используют методы `.head(n)` и `.tail(n)`.
* `.head()` возвращает первые `n` строк (по умолчанию 5).
* `.tail()` возвращает последние `n` строк.
Это помогает быстро убедиться, что данные загрузились корректно.

```python
df.head()
```

## 2. Базовые характеристики датасета

### Атрибут `.shape`
У любого `DataFrame` есть атрибут `.shape`. Обратите внимание: это **атрибут**, а не метод, поэтому он вызывается без круглых скобок в конце. Он возвращает кортеж из двух чисел: `(количество_строк, количество_столбцов)`.

```python
df.shape
```

### Атрибут `.columns`
Тоже атрибут, возвращающий специальный объект (похожий на список) с названиями всех колонок в таблице. Удобно использовать, если колонок много и они не помещаются на экран.

```python
df.columns
```

### Метод `.info()`
Этот метод выдает «паспорт» датасета. Что мы из него узнаем:
1. Индексы датасета (сколько всего записей).
2. Название каждой колонки.
3. Количество Non-Null (непустых) значений в каждой колонке. Сразу видно, где есть пропуски.
4. Тип данных (Dtype) каждой колонки (например, `int64` — целые числа, `float64` — дробные, `object` — строки).
5. Объем оперативной памяти, который занимает таблица.

```python
df.info()
```

### Метод `.describe()`
Генерирует описательную статистику. По умолчанию он берет только **числовые** колонки и считает для них:
- `count` (количество непустых),
- `mean` (среднее арифметическое),
- `std` (стандартное отклонение, разброс),
- `min` и `max`,
- `25%`, `50%` (медиана), `75%` — квантили.

```python
df.describe()
```

## 3. Индексация и выбор данных

Обращение к одной колонке происходит через квадратные скобки по ее названию. В результате мы получаем не DataFrame, а объект типа **Series** (одномерный массив с индексами).

```python
df['track_name'].head()
```

Если нужно выбрать сразу несколько колонок, мы передаем в квадратные скобки **список** их названий. Обратите внимание на двойные скобки `[[...]]`: внешние — для индексации датафрейма, внутренние — это сам список строк.

```python
ds = df[['track_id', 'track_name', 'track_artist', 'track_popularity',
       'track_album_id', 'track_album_name', 'track_album_release_date',
       'playlist_name', 'playlist_id', 'playlist_genre', 'playlist_subgenre',
       'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'liveness', 'valence', 'tempo',
       'duration_ms']].head()
ds
```

### `.loc` и `.iloc`

Это два главных свойства для продвинутой индексации:
* `.iloc` (integer location) — ищет по **порядковому номеру** (от 0 до n-1). Работает ровно так же, как стандартные срезы в Python. *Правая граница не включается*.
* `.loc` (location) — ищет по **метке** (имени индекса/колонки). Важнейшее отличие от обычного Python: при использовании срезов через `.loc` *правая граница ВКЛЮЧАЕТСЯ* в результат.

```python
# .iloc: выбираем строки с 10 по 14 (15-я не включится)
df.iloc[10:15]
```

```python
# .loc: выбираем строки с индексами от 10 до 14 ВКЛЮЧИТЕЛЬНО,
# и указываем конкретные колонки
df.loc[10:14, ['track_name', 'track_popularity']]
```

## 4. Логическое индексирование (Фильтрация)

В Pandas фильтрация работает через «булевы маски». Выражение вида `df['track_popularity'] > 95` возвращает не сами строки, а объект Series из True/False для каждой строки. Передавая эту «маску» в `df[...]`, мы оставляем только те строки, где стоит True.

```python
# Треки с популярностью больше 95
df[df['track_popularity'] > 95].head()
```

### Сложные условия
Если нужно несколько условий, обычные `and` и `or` в Python не сработают (они пытаются оценить весь массив целиком, что вызывает ошибку `ValueError`).
Вместо них используются побитовые операторы:
* `&` (логическое И)
* `|` (логическое ИЛИ)
* `~` (логическое отрицание НЕ)

**Важно:** каждое условие обязательно нужно оборачивать в круглые скобки `()`, так как приоритет побитовых операторов выше, чем у операторов сравнения (`>`, `<` и т.д.).

```python
df[(df['playlist_genre'] == 'pop') | (df['danceability'] > 0.9)].head()
```

```python
df[(df['track_artist'] == 'Ed Sheeran') & (df['track_popularity'] > 85)][['track_name', 'track_popularity']].shape

ds = df[(df['track_artist'] == 'Ed Sheeran') & (df['track_popularity'] > 85)]
ds[df['track_name'] == 'Shape of You']
```

## 5. Группировка данных (.groupby)

Группировка (`.groupby`) реализует концепцию **Split-Apply-Combine**:
1. **Split**: таблица разбивается на куски (группы) по уникальным значениям выбранного столбца.
2. **Apply**: к каждому куску применяется агрегирующая функция (среднее, сумма, минимум и т.д.).
3. **Combine**: результаты склеиваются в новую таблицу, где индексами становятся уникальные значения столбца, по которому мы группировали.

Например, сгруппируем по жанру и посмотрим среднюю популярность:

```python
df.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=True).tail(3)
```

### Метод `.agg()`
Часто нужно посчитать не одну метрику, а сразу несколько, причем для разных колонок — разные функции. Метод `.agg()` принимает словарь, где ключи — это названия колонок, а значения — функции (в виде строк или списков строк), которые нужно к ним применить.

```python
df.groupby('playlist_genre').agg({
    'track_popularity': ['mean', 'count'],
    'danceability': 'mean',
    'tempo': 'median'
})
```

**Задание 2.** Сгруппируйте датасет по поджанру (`playlist_subgenre`). Для каждого поджанра найдите максимальную энергичность (`energy`) и минимальную акустичность (`acousticness`). Отсортируйте результат по убыванию максимальной энергичности.

```python
df.groupby('playlist_subgenre').agg({
    'energy': 'max',
    'acousticness': 'min'
}).sort_values(by='energy', ascending=False)
```

## 6. Создание признаков и метод `.apply()`

Мы можем легко создавать новые колонки, используя арифметические операции над старыми. Это называется векторизованными операциями — pandas применяет математику сразу ко всему столбцу, не используя циклы (что очень быстро).

```python
# Переводим миллисекунды в минуты
df['duration_min'] = df['duration_ms'] / 60000
df[['track_name', 'duration_ms', 'duration_min']].head()
```

Если логика сложная и обычными арифметическими операциями не обойтись, мы пишем свою Python-функцию и применяем её к колонке с помощью `.apply()`.

Метод `.apply()` проходит по каждой ячейке в колонке, передает ее значение в функцию и записывает результат в новую ячейку. Часто `.apply()` используют в связке с анонимными лямбда-функциями (`lambda x: ...`), но для читаемости можно передавать и обычные `def`.

```python
def get_popularity_status(pop):
    if pop > 80:
        return 'Hit'
    elif pop > 50:
        return 'Average'
    else:
        return 'Low'

df['popularity_status'] = df['track_popularity'].apply(get_popularity_status)

# Метод .value_counts() считает количество каждого уникального значения в столбце
df['popularity_status'].value_counts()
```

## 7. Пропущенные значения (NaN)

В реальных датасетах всегда есть пропуски (NaN - Not a Number).
Метод `.isna()` (или его синоним `.isnull()`) возвращает DataFrame такого же размера, но состоящий из True/False, где True стоит на месте пропуска.

Если применить к этому результату `.sum()`, pandas сложит True (считая их за 1) по каждой колонке и покажет общее число пропусков.

```python
df.isna().sum()
```

### Метод `.dropna()` и `.fillna()`
Есть два основных пути работы с пропусками:
1. Заполнить их чем-то осмысленным с помощью `.fillna(значение)` (например, нулями или средним).
2. Выбросить строки с пропусками с помощью `.dropna()`.

В нашем случае пропусков всего по 5 штук в трех текстовых колонках при общем объеме в 32000 строк. Потерять 5 строк — не страшно, поэтому смело дропаем.

```python
df = df.dropna()
df.isna().sum()
```
